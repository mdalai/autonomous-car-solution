```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os.path
import numpy as np
import scipy.misc
```

```python
import tensorflow as tf
```

```python
# Image Plot helper function
def plot_img(img, transformed_img, method_name, gray=False, original_gray=False):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 9))
    #f.tight_layout()
    if original_gray:
        ax1.imshow(img,cmap='gray')
    else:
        ax1.imshow(img)
    ax1.set_title('Original Image')
    if gray:
        ax2.imshow(transformed_img, cmap='gray')
    else:
        ax2.imshow(transformed_img)
    ax2.set_title(method_name)
    #plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
sess = tf.Session()
saver = tf.train.import_meta_graph('data/model_2c/model_2c.meta')
saver.restore(sess,tf.train.latest_checkpoint('data/model_2c'))
graph = tf.get_default_graph()
labels = graph.get_tensor_by_name("labels:0")
logits_2d = graph.get_tensor_by_name("logits_2d:0")
labels_2d = graph.get_tensor_by_name("labels_2d:0")
loss = graph.get_tensor_by_name("loss:0")
image_input = graph.get_tensor_by_name("image_input:0")
keep_prob = graph.get_tensor_by_name("keep_prob:0")
lr = graph.get_tensor_by_name("learning_rate:0")    
```

```python
im_softmax = sess.run([tf.nn.softmax(logits_2d)],{keep_prob: 1.0, image_input: [img2]})
```

```python
# get the NOT Background softmax value and reshape to the image size
im_softmax = im_softmax[0][:,1].reshape(160, 576)
```

```python
segmentation = (im_softmax > 0.5).reshape(image_shape[0], image_shape[1], 1)
```

## Understand Data

```python
import os, os.path
from glob import glob
import re
import numpy as np

import scipy.misc
```

```python
DATA_DIR = 'data/data_road/training'
```

```python
image_paths = glob(os.path.join(DATA_DIR, 'image_2', '*.png'))
label_paths = {
            re.sub(r'_(lane|road)_', '_', os.path.basename(path)): path
            for path in glob(os.path.join(DATA_DIR, 'gt_image_2', '*_road_*.png'))}
```

```python
print("Total training images: ",len(image_paths))
print("Total training lables: ",len(label_paths))
```

```python
img_path = image_paths[0]
train_img = mpimg.imread(img_path)
label_img = mpimg.imread(label_paths[os.path.basename(img_path)])

fig = plt.figure(figsize=(17,15))
plt.subplot(121)
plt.imshow(train_img)
plt.title('Train')
plt.subplot(122)
plt.imshow(label_img)
plt.title('Label')
```

```python
# read images
img_path = image_paths[100]
train_img = scipy.misc.imread(img_path)
label_img = scipy.misc.imread(label_paths[os.path.basename(img_path)])
# resize images
train_img = scipy.misc.imresize(train_img, (160, 576))
label_img = scipy.misc.imresize(label_img, (160, 576))

fig = plt.figure(figsize=(17,15))
plt.subplot(121)
plt.imshow(train_img)
plt.title('Train')
plt.subplot(122)
plt.imshow(label_img)
plt.title('Label')
```

```python
# Red is background color
background_color = np.array([255, 0, 0])
```

```python
# picking up the channel value and compare it with defined bg color
gt_bg = np.all(label_img == background_color, axis=2)
gt_bg.shape
```

```python
# change the shape to one channel
gt_bg = gt_bg.reshape(*gt_bg.shape, 1)
gt_bg.shape              
```

```python
# concatenate to generate 2 channels, yes/no
gt_image = np.concatenate((gt_bg, np.invert(gt_bg)), axis=2)
gt_image.shape
```
